#seasons for the data ->slower in summer than in winter 

. spring 3-5

. summer 6-8

. fall  9-11

. winter 12-2

```python
df5['race_month']=df5['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df5.head()
```

```python
df5['race_season']=df5['race_month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df5.tail(25)
```

```python
df5.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

#Hence it makes sense that lot of marathons happen in winters as it is a suitable weather 


```python
#50 miler only

df5.query('race_length=="50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
